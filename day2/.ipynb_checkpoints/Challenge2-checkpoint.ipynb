```python
from pathlib import Path

input_file = Path(".") / "example1.txt"

# check if id is composed of two duplicated sequence
def is_duplicated(id: str) -> bool:
    length = len(id)
    if length % 2 == 1: # odd lengths are never duplicate
        return False
    half_length = int(length / 2)
    
    return id[0:half_length] == id[half_length:length]

# check if id is a composition of repeated digits
def is_repeated(id: str) -> bool:
    return re.match(r"^(\d+?)\1+$", id) != None

sum_of_duplicated_ids = 0
sum_of_repeated_ids = 0
with input_file.open(mode="r", encoding="utf-8") as file:
    for line in file:
        for id_range in line.split(","):
            if len(id_range.strip()) > 0:
                id_from, id_to = id_range.split("-")
                for id_ in range(int(id_from), int(id_to) + 1):
                    if is_duplicated(str(id_)):
                        #print(f"{id_} ")
                        sum_of_duplicated_ids += id_
                    if is_repeated(str(id_)):
                        #print(f"{id_} ")
                        sum_of_repeated_ids += id_
                
print(f"Part1 answer: {sum_of_duplicated_ids}")
print(f"Part2 answer: {sum_of_repeated_ids}")


```
