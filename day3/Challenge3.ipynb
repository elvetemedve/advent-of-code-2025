```python
from pathlib import Path

input_file = Path(".") / "input.txt"

def max_n_joltage(bank: str, n: int) -> int:
    bank_length = len(bank)
    start_position = 0
    max_digits = ""

    for digits_remaining in range(0, n)[::-1]:
        search_from = start_position
        search_to = bank_length - digits_remaining
        max_digit = max(list(bank[search_from:search_to]))
        max_digits += max_digit
        start_position = bank[start_position:].index(max_digit) + start_position + 1

    #print(f"Bank {bank}, joltage {max_digits}")
    return int(max_digits)

total_output_joltage1 = 0
total_output_joltage2 = 0
with input_file.open(mode="r", encoding="utf-8") as file:
    for line in file:
        total_output_joltage1 += max_n_joltage(line.strip(), 2)
        total_output_joltage2 += max_n_joltage(line.strip(), 12)
                
print(f"Part1 answer: {total_output_joltage1}")
print(f"Part2 answer: {total_output_joltage2}")


```
