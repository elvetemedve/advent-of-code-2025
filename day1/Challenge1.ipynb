```python
from pathlib import Path
import re
import math

current_position = previous_position = 50
count_of_zero_endings = 0
count_of_zero_crossings = 0
input_path = Path(".") / "challenge1.txt"

with input_path.open(mode="r", encoding="utf-8") as file:
    for line in file:
        matches = re.match(r"(R|L)(\d+)", line)
        turns = int(matches.group(2))
        if matches.group(1) == "L":
            turns *= -1

        relative_position = current_position + turns
        current_position = relative_position % 100
        if relative_position <= 0:
            overflow = math.trunc((abs(relative_position / 100)) + (1 if previous_position > 0 else 0))
            #print(f"\nrelative={relative_position} overflow={overflow}")
            relative_position = current_position
            count_of_zero_crossings += overflow
        
        count_of_zero_crossings += math.trunc(relative_position / 100)
        
        if current_position == 0:
            count_of_zero_endings += 1        

        #print(f"Position: {previous_position} Turns: {turns} Crossings: {count_of_zero_crossings}")

        previous_position = current_position

print(f"Part1 answer: {count_of_zero_endings}")
print(f"Part2 answer: {count_of_zero_crossings}")
```
