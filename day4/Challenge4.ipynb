```python
from pathlib import Path
import copy

input_file = Path(".") / "input.txt"

# lambda function for reading matrix elements without index out of bounds
matrix_element = lambda m, x, y: m[x][y] if 0 <= x < len(m) and 0 <= y < len(m[x]) else 0

def count_movable_papers(matrix) -> tuple:
    movable_count = 0
    updated_matrix = copy.deepcopy(matrix)
    get = lambda x, y: matrix_element(matrix, x, y)
    for i, row in enumerate(matrix):
        for j, element in enumerate(row):
            if element == 1:
                adjacent_count = get(i-1,j) + get(i+1,j) + get(i,j-1) + get(i,j+1) + \
                    get(i-1,j-1) + get(i+1,j-1) + get(i-1,j+1) + get(i+1,j+1)
                if adjacent_count < 4:
                    movable_count += 1
                    updated_matrix[i][j] = 0

    return (movable_count, updated_matrix)

input_matrix = [] # elemnts are 1 where there is a paper roll, otherwise 0
with input_file.open(mode="r", encoding="utf-8") as file:
    for line in file:
        parsed_line = re.sub(r"[^\.@]*", "", line)
        row = []
        for char in parsed_line:
            row.append(1 if char == '@' else 0)
        input_matrix.append(row)

    matrix = input_matrix
    moveable_paper_rolls_count = 0
    total_moveable_paper_rolls_count = 0
    while True:
        count, matrix = count_movable_papers(matrix)
        #print(f"Removed paper rolls count: {count}")
        if moveable_paper_rolls_count == 0:
            moveable_paper_rolls_count = count
        total_moveable_paper_rolls_count += count
        if count == 0:
            break
                
print(f"Part1 answer: {moveable_paper_rolls_count}")
print(f"Part2 answer: {total_moveable_paper_rolls_count}")



```
