```python
from pathlib import Path
from itertools import combinations

input_file = Path(".") / "example1_p1-50.txt"

def find_greatest_area(points: list) -> int:
    max_area = 0
    for point1, point2 in list(combinations(points, 2)):
        x1, y1 = point1
        x2, y2 = point2
        a = abs(x1-x2) + 1
        b = abs(y1-y2) + 1
        area = a * b
        if area > max_area:
            max_area = area
    return max_area

points = []
with input_file.open(mode="r", encoding="utf-8") as file:
    for line in file:
        points.append(list(map(int,line.strip().split(","))))

find_greatest_area(points)
```
